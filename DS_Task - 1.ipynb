##  Step - 3 :  Data Visualization

### 3.1 Showcasing Global Population Over Year of 2013 to 2023

```python
# Filter the years of interest (2013 - 2023)
years = [str(year) for year in range(2013, 2024)]
global_population = data[years].sum()
global_population_df = global_population.reset_index()
global_population_df.columns = ['Year', 'Global Population']
global_population_df['Year'] = global_population_df['Year'].astype(int)
```

```python
# Plot the global population trend 
plt.figure(figsize=(8, 5))
plt.plot(global_population_df['Year'], global_population_df['Global Population'], marker='o', linestyle='-', color='b')
plt.title('Global Population Growth (2013 - 2023)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Global Population', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### 3.2 The bar chart illustrating the population distribution across global regions from 2013 & 2023

```python
# Mapping of regions if not present in the dataset 
region_mapping = {
    'India': 'Asia', 'China': 'Asia', 'Japan': 'Asia', 'Germany': 'Europe', 'France': 'Europe',
    'Nigeria': 'Africa', 'South Africa': 'Africa', 'Brazil': 'South America',
    'United States': 'North America', 'Canada': 'North America', 'Australia': 'Oceania'
}
data['Region'] = data['Country Name'].map(region_mapping)
years = ['2013', '2023']
regional_population = data.groupby('Region')[years].sum()

# Calculate the growth percentage for each region
regional_population['Growth (%)'] = ((regional_population['2023'] - regional_population['2013']) / regional_population['2013']) * 100
regional_population = regional_population.reset_index()
```

```python
# Create the Vertical bar plot 
x = np.arange(len(regional_population['Region']))  # X-axis positions
width = 0.25  
plt.figure(figsize=(7, 5))
plt.bar(x - width/2, regional_population['2013'], width, label='2013', color='skyblue')
plt.bar(x + width/2, regional_population['2023'], width, label='2023', color='orange')
plt.title('                Population Growth by Region (2013 vs 2023)', fontsize=12)
plt.xlabel('Region', fontsize=12)
plt.ylabel('Population', fontsize=12)
plt.xticks(x, regional_population['Region'], rotation=45)
plt.legend()

# finding growth percentages above bars
for i in range(len(x)):
    plt.text(x[i], regional_population['2023'][i] + 0.05 * regional_population['2023'].max(),
             f"{regional_population['Growth (%)'][i]:.2f}%", ha='center', fontsize=10)
plt.tight_layout()
plt.show()
```

### 3.3 Selecting a Specific Country & Visualize Population Growth Over Time 0f 10 years

```python
# Define the country for analysis
country = 'United States'
country_data = data[data['Country Name'] == country]
years = [str(year) for year in range(2013, 2024)]
population_trend = country_data[years].melt(var_name='Year', value_name='Population')
population_trend['Year'] = population_trend['Year'].astype(int)
```

```python
# Plot the population trend 
plt.figure(figsize=(7, 5))
plt.plot(population_trend['Year'], population_trend['Population'], marker='o', linestyle='-', color='green')
plt.title(f'Population Growth Trend: {country} (2013–2023)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Population', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

# & showing the highest and lowest points
max_population = population_trend['Population'].max()
min_population = population_trend['Population'].min()
max_year = population_trend.loc[population_trend['Population'].idxmax(), 'Year']
min_year = population_trend.loc[population_trend['Population'].idxmin(), 'Year']
plt.annotate(f"Highest: {max_population:,.0f}", xy=(max_year, max_population), 
             xytext=(max_year - 2, max_population + 5e6), 
             arrowprops=dict(facecolor='black', arrowstyle='->'))
plt.annotate(f"Lowest: {min_population:,.0f}", xy=(min_year, min_population), 
             xytext=(min_year + 2, min_population - 5e6), 
             arrowprops=dict(facecolor='black', arrowstyle='->'))
plt.tight_layout()
plt.show()
```

### 3.4 Highlighting countries with the highest or lowest growth (absolute or percentage-wise) in a specific year of 2023.

```python
# Calculate population growth (absolute) between 2022 and 2023 for all countries
years_growth = ['2022', '2023']
growth_data = data[['Country Name'] + years_growth].dropna()
growth_data['Absolute Growth'] = growth_data['2023'] - growth_data['2022']
growth_data['Percentage Growth'] = ((growth_data['2023'] - growth_data['2022']) / growth_data['2022']) * 100
top_growth_countries = growth_data.nlargest(10, 'Absolute Growth')
least_growth_countries = growth_data.nsmallest(10, 'Absolute Growth')
```

### 3.4.1 Visualizing Highest Population Growth of 10 countries in 2023

```python
# Top 10 countries population growth Visualization using hosizantal bar chart
plt.figure(figsize=(8, 4))
plt.barh(top_growth_countries['Country Name'], top_growth_countries['Absolute Growth'], color='green')
plt.title('Top 10 Countries of Population Growth (2023)', fontsize=14)
plt.xlabel('Absolute Population Growth', fontsize=12)
plt.ylabel('Country', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### 3.4.2 Visualizing lowest Population Growth of 10 countries in 2023

```python
# Least Top 10 countries population growth Visualization using hosizantal bar chart
plt.figure(figsize=(8, 4))
plt.barh(least_growth_countries['Country Name'], least_growth_countries['Absolute Growth'], color='red')
plt.title('Top 10 Countries by Least Population Growth (2023)', fontsize=14)
plt.xlabel('Absolute Population Growth', fontsize=12)
plt.ylabel('Country', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### 3.5 Demonstrating population growth rates for 10 countries across 10 years using Heatmap

```python
# Select 10 random countries for analysis & growth rates
selected_countries = data[['Country Name'] + [str(year) for year in range(2013, 2024)]].dropna().sample(10, random_state=42)
growth_rates = selected_countries.set_index('Country Name').pct_change(axis=1) * 100
growth_rates = growth_rates.dropna(axis=1) 

# Create a heatmap
plt.figure(figsize=(8, 5))
sns.heatmap(growth_rates, annot=True, fmt=".1f", cmap="coolwarm", linewidths=0.5, cbar_kws={'label': 'Growth Rate (%)'})
plt.title('Year-over-Year Population Growth Rates (2013–2023)', fontsize=12)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Country', fontsize=12)
plt.tight_layout()
plt.show()
```

### 3.6 Stacked bar chart showing continent-wise population distribution over 10 years.

```python
# Map countries to continents 
continent_mapping = {
    'China': 'Asia',
    'India': 'Asia',
    'United States': 'North America',
    'Brazil': 'South America',
    'Germany': 'Europe',
    'Nigeria': 'Africa',
    'South Africa': 'Africa',
    'Australia': 'Oceania',
}
data['Continent'] = data['Country Name'].map(continent_mapping)
years = [str(year) for year in range(2013, 2024)]
continent_data = data[['Continent'] + years].dropna()
continent_population = continent_data.groupby('Continent')[years].sum()
continent_population = continent_population.T
```

```python
# Plot stacked bar chart
continent_population.plot(kind='bar', stacked=True, figsize=(8, 5), colormap='tab20')
plt.title('Continent-Wise Population Distribution (2013–2023)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Population', fontsize=12)
plt.legend(title='Continent', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### 3.7 Illustrates the global population growth timeline from 2013 to 2023 and Milestones such as the COVID-19 pandemic (2019) and post-pandemic recovery (2023) are annotated 

```python
# global growth population in 10 years with significant milestones
years = [str(year) for year in range(2013, 2024)]
global_population = data[years].sum().reset_index()
global_population.columns = ['Year', 'Population']
global_population['Year'] = global_population['Year'].astype(int)

# Create a step chart to visualize significant milestones with historical context
plt.figure(figsize=(7, 5))
plt.step(global_population['Year'], global_population['Population'], where='mid', color='blue', linestyle='-', label='Global Population')
plt.title('Global Population Growth Timeline (2013–2023)', fontsize=12)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Global Population', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
milestones = [
    (2019, 'COVID-19 Pandemic'), 
    (2023, 'Post-pandemic Recovery')
]

for year, event in milestones:
    population = global_population.loc[global_population['Year'] == year, 'Population'].values[0]
    plt.annotate(f"{event}\n{population:,.0f}", xy=(year, population), 
                 xytext=(year - 2, population + 1e9), 
                 arrowprops=dict(facecolor='red', arrowstyle='->'), fontsize=10)
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()
```

### 3.8  Highlighting which continents dominate the global population in 2023

```python
# Grouping by continent and percentage contributions population for 2023 
continent_population_2023 = data.groupby('Region')['2023'].sum()
total_population_2023 = continent_population_2023.sum()
continent_population_2023_percentage = (continent_population_2023 / total_population_2023) * 100
continent_population_df = continent_population_2023_percentage.reset_index()
continent_population_df.columns = ['Continent', 'Population Percentage']

# Plot a pie chart
plt.figure(figsize=(6, 4))
plt.pie(
    continent_population_df['Population Percentage'], 
    labels=continent_population_df['Continent'], 
    autopct='%1.1f%%', 
    startangle=140, 
    colors=plt.cm.Paired.colors
)
plt.title('Continent-wise Population Contribution (2023)', fontsize=13)
plt.tight_layout()
plt.show()
```

### 3.9 Exracting the 20-Year Population Growth of Top vs Least Growing Asian Countries

```python
# 20-year of analysis for asian countries
start_year = '2003'
end_year = '2023'
asian_countries = ['India', 'China', 'Japan', 'Indonesia', 'Pakistan', 'Bangladesh', 
                   'Vietnam', 'Philippines', 'Thailand', 'Malaysia']  
asian_data = data[data['Country Name'].isin(asian_countries)]
asian_data['Growth (2003-2023)'] = (asian_data[end_year] - asian_data[start_year]) / asian_data[start_year] * 100
top_asian_growth = asian_data[['Country Name', 'Growth (2003-2023)']].sort_values(by='Growth (2003-2023)', ascending=False).head(5)
least_asian_growth = asian_data[['Country Name', 'Growth (2003-2023)']].sort_values(by='Growth (2003-2023)', ascending=True).head(5)
```

```python
# Visualization of 20 years Top vs Least Growing Asian Countries
plt.figure(figsize=(7, 5))
sns.barplot(data=top_asian_growth, x='Growth (2003-2023)', y='Country Name', palette='viridis', label='Top Growth')
sns.barplot(data=least_asian_growth, x='Growth (2003-2023)', y='Country Name', palette='magma', label='Least Growth', alpha=0.7)
plt.title('20-Year Population Growth: Top vs Least Growing Asian Countries', fontsize=12)
plt.xlabel('Growth Rate (%) (2003-2023)', fontsize=12)
plt.ylabel('Country', fontsize=12)
plt.legend(['Top Growing', 'Least Growing'])
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### 3.10 The boxplot represents the variability of country populations across different income groups.

```python
# Map income groups to countries
if 'Income Group' not in data.columns:
    income_group_mapping = {
        'United States': 'High-income', 
        'India': 'Low-income', 
        'China': 'Middle-income',
        'Nigeria': 'Low-income', 
        'Germany': 'High-income',
        'Brazil': 'Middle-income'
    }
    data['Income Group'] = data['Country Name'].map(income_group_mapping)
income_group_population = data[['Country Name', 'Income Group', '2023']].dropna()
income_group_population.rename(columns={'2023': 'Population (2023)'}, inplace=True)
```

```python
# Create a boxplot population distribution of income group
plt.figure(figsize=(7, 5))
sns.boxplot(
    x='Income Group', 
    y='Population (2023)', 
    data=income_group_population, 
    palette='Set2'
)
plt.title('Population Distribution by Income Group (2023)', fontsize=14)
plt.xlabel('Income Group', fontsize=12)
plt.ylabel('Population (2023)', fontsize=12)
plt.yscale('log')
plt.tight_layout()
plt.show()
```
